# Part 2: Data Cleaning and Manipulation
# (Pandas)


```python
1. Calculate the number of missing values in each column.
Step: Use .isnull() followed by .sum() .
2. Check if there are any duplicate rows in the DataFrame.
Step: Use .duplicated() and .sum() .
3. Rename the SuccessfulOrder column to OrderSuccess .
Step: Use the .rename() method with the columns parameter.
4. Create a new column named ItemsCount which is the sum of ToppingsCount and
ScoopsCount .
Step: Add the two columns together.
5. Calculate the average TotalCost for each PaymentMethod .
Step: Use the .groupby() method on PaymentMethod and then apply
.mean() to TotalCost .
```

```python
# 1 .Calculating total Null Values
null_value=df.isnull().sum()
print(null_value)
```

```python
# 2. Checking Duplicate Rows
duplicate_value=df.duplicated().sum()
print(duplicate_value)
```

```python
# 3.Renaming the columns
df=df.rename(columns={"SuccessfulOrder":"OrderSuccess"})
df.head()
```

```python
# 4.New Column Creation
df["ItemsCount"]=df["ToppingsCount"]+df["ScoopsCount"]
df.head()
```

```python
# 5.Average Total cost of PayementMethod
avg_total_cost=df.groupby("PaymentMethod")["TotalCost"].mean()
print(avg_total_cost)
```

# Part 3: Basic Exploratory Data Analysis (EDA) and NumPy

```python
1. Calculate the mean, median, and standard deviation of CustomerAge .
Step: Use .mean() , .median() , and .std() on the CustomerAge column.
2. Find the most common PaymentMethod and FlavorCategory .
Step: Use the .mode() method.
3. Count the number of successful and unsuccessful orders.
Step: Use the .value_counts() method on the OrderSuccess column.
4. Use NumPy to calculate the percentage of successful orders.
Step: Calculate the sum of successful orders and divide by the total number of
records.
5. Find the maximum TotalCost and the minimum TotalCost from the dataset.
Step: Use .max() and .min() on the TotalCost column
```

```python
# 1.Calculating the mean,std,median
mean=df["CustomerAge"].mean()
std=df["CustomerAge"].std()
median=df["CustomerAge"].median()
print("Mean:",mean)
print(f"Std:{std:.4f}")
print("Median:",median)
```

```python
# 2.most common PaymentMethod and FlavorCategory
common_payment_method=df["PaymentMethod"].mode()
common_flavor_category=df["FlavorCategory"].mode()
print(common_payment_method)
print(common_flavor_category)
```

```python
# 3.Count Successful and Unsuccessful order
order_made=df["OrderSuccess"].value_counts()
print(order_made)
print("There is no indication of successful order in the (OrderSuccess) column")
```

```python
df["OrderSuccess"].nunique()
```

```python
# 4.Percentage of Successful order
import numpy as np
successfull_order=df["OrderSuccess"].sum()
total_number_records=5000
percentage_successfull_order=successfull_order/total_number_records
print("Percentage :",percentage_successfull_order)
```

```python
# 5.Max & Min of totalcost
max_cost=df["TotalCost"].max()
min_cost=df["TotalCost"].min()
print(f"Max Cost:{max_cost:.4f}")
print(f"Min Cost:{min_cost:4f}")
```

```python
# HistPlot Using Matplotlib
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.hist(df["CustomerAge"],bins=5,color="orange",edgecolor="green",label="CustomerAge")
plt.title("Hist Plot",color="darkgreen",fontweight="bold",fontsize=15)
plt.xlabel("Customer Age",color="blue",fontweight=13)
plt.ylabel("Values",color="blue",fontweight=13)
plt.legend()
plt.grid(True,alpha=0.2)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
# 2.Seaborn
import seaborn as sns
sns.countplot(x="PaymentMethod",data=df,hue="PaymentMethod",palette="coolwarm",saturation=1,edgecolor="black",dodge=True)
plt.title("Count Plot",color="darkgreen",fontweight="bold",fontsize=15)
plt.xlabel("PaymentMethod",color="blue",fontweight=13)
plt.ylabel("Values",color="blue",fontweight=13)
plt.grid(True,alpha=0.2)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
mean_value=df.groupby("PaymentMethod")["TotalCost"].mean()
mean_value
```

```python
# 7. Create a bar chart showing the average TotalCost per PaymentMethod .
# Step: Group the data by PaymentMethod , calculate the mean TotalCost , and
# then use sns.barplot() .

mean_value=df.groupby("PaymentMethod",as_index=False)["TotalCost"].mean()
plt.figure(figsize=(10,8))
sns.barplot(x="PaymentMethod",y="TotalCost",data=mean_value,hue="PaymentMethod",palette="Set1",dodge=True,width=0.5)
plt.title("Bar Plot",color="darkgreen",fontweight="bold",fontsize=15)
plt.xlabel("PaymentMethod",color="black",fontweight=13)
plt.ylabel("mean_value",color="black",fontweight=13)
plt.grid(True,alpha=0.2)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 8. Visualize the correlation matrix of the numerical columns using a heatmap.
# Step: Calculate the correlation matrix using .corr() and then use
# sns.heatmap() .
corr=df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True,cmap="viridis",linewidth=0.5)
plt.title("Correlation of matrix of the numerical columns")
plt.show()
```
