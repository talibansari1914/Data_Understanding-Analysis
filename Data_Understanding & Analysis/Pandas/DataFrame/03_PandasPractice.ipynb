```python
import pandas as pd
filename="product_batch_with_nan.csv"
chunk_size=50
revenue=0
for chunk in pd.read_csv(filename,chunksize=chunk_size):
    chunk["Total"]=chunk["Price"]*chunk["Quantity"]
    revenue+=chunk["Total"].sum()
    print("Current revenue is:",revenue)
print("Total revenue is:",revenue)
```
