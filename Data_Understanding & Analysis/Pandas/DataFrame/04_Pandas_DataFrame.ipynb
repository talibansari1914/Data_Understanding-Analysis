```python
import pandas as pd

df1=pd.DataFrame({"ID": [1, 2, 3, 4],
    "Name": ["Alice", "Bob", "Charlie", "David"]})
df2=pd.DataFrame({
    "ID": [3, 4, 5, 6],
    "City": ["New York", "Chicago", "Houston", "Phoenix"]
})
inner=pd.merge(df1,df2,how="inner")
print(inner)
```

```python
left=pd.merge(df1,df2,how="left")
print(left)
```

```python
right=pd.merge(df1,df2,how="right")
print(right)
```

```python
outer=pd.merge(df1,df2,how="outer")
print(outer)
```

```python
df3=pd.concat([df1,df2])
print(df3)
```

```python
import pandas as pd
students = pd.DataFrame({
    "Roll_No": [101, 102, 103, 104],
    "Name": ["Aman", "Riya", "Vikram", "Sara"]
})
marks = pd.DataFrame({
    "Roll_No": [103, 104, 105, 106],
    "Marks": [88, 92, 75, 60]
})
#*****************  JOIN  ****************========************====================
inner_1=pd.merge(students,marks,how="inner")
print(inner_1)
```

```python
left_2=pd.merge(students,marks,how="left")
print(left_2)
```

```python
right_3=pd.merge(students,marks,how="right")
print(right_3)
```

```python
outer_4=pd.merge(students,marks,how="outer")
print(outer_4)
```

```python
df2=pd.concat([students,marks])
print(df2)
```
