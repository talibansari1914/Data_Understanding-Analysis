```python
import pandas as pd
students_data = {
    "Student_ID": [101, 102, 103, 104, 105, 106, 107],
    "Name": ["Aman", "Bhavna", "Chirag", "Divya", "Eshan", "Fatima", "Gaurav"],
    "Maths": [87, 78, 90, 66, 92, 81, 74],
    "Science": [82, 69, 85, 70, 89, 76, 80],
    "English": [75, 88, 79, 91, 73, 85, 68],
    "Attendance(%)": [95, 88, 90, 92, 85, 93, 87]
}
data_df=pd.DataFrame(students_data)
data_df
```

```python
data_df1=data_df.groupby(["Name","Maths"])["Science"].sum().reset_index()
data_df1
```

```python
data_df1=data_df.groupby(["Name","Maths"])[["Science","English","Student_ID"]].sum().reset_index()
data_df1
```

```python
data_df1=data_df.groupby(["Name","Maths"])[["Science","English","Student_ID"]].sum().unstack()
data_df1
```

```python
data_df1=data_df.groupby(["Name","Maths"])[["Science"]].sum().unstack()
data_df1
```

```python
data_df1=data_df.groupby(["Name","Maths"])[["Science"]].max().unstack()
data_df1
```
