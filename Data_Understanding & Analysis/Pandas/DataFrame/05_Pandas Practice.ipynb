```python
import pandas as pd
import numpy as np

# Create our primary sales data file
data_sales = {
    'OrderID': [101, 102, 103, 104, 105],
    'Date': pd.to_datetime(['2025-08-01', '2025-08-01', '2025-08-02', '2025-08-03', '2025-08-04']),
    'Salesperson': ['Priya', 'Amit', 'Priya', 'Neha', 'Rohit'],
    'Item_Code': ['V_01', 'C_02', 'V_01', 'P_03', 'M_04'],
    'Quantity': [5, 3, 4, 6, 3],
    'Price_INR': [120, 150, 120, 180, 160]
}
sales_df = pd.DataFrame(data_sales)
data_items = {
    'Item_Code': ['V_01', 'C_02', 'P_03', 'M_04', 'S_05'],
    'Item_Name': ['Vanilla', 'Chocolate', 'Pista', 'Mango', 'Strawberry'],
    'Category': ['Standard', 'Standard', 'Premium', 'Premium', 'Standard']
}
items_df = pd.DataFrame(data_items)
```

```python
inner=sales_df.merge(items_df,on="Item_Code",how="inner")
inner
```

```python
outer=sales_df.merge(items_df,on="Item_Code",how="outer")
outer
```

```python
left=sales_df.merge(items_df,on="Item_Code",how="left")
left
```

```python
right=sales_df.merge(items_df,on="Item_Code",how="right")
right
```
