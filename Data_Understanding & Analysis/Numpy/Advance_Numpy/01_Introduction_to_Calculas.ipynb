```python
# This script defines and visualizes key functions used in mathematics,
# statistics, and machine learning, with explanations suitable for a non-technical audience.

import numpy as np
import matplotlib.pyplot as plt

# Set plotting format for better resolution
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')

# --- 1. Define the Domain (x-values) ---
# We define a range of input values (x) to plot our functions over.
xx = np.linspace(-10, 10, 400)
# Separate domain for functions undefined for x <= 0 (Logarithm, Square Root)
xx_pos = np.linspace(0.1, 10, 400)


# --- 2. Define the Functions (Ordered from Basic to Specialized) ---

# --- Basic Mathematical Functions (1-4) ---

def linear_function(x):
    """1. Simple straight line: y = 2x + 1. Used in basic prediction models."""
    return 2 * x + 1

def quadratic_function(x):
    """2. Parabola: y = x^2 - 3. Creates a 'U' shape, essential for error calculation."""
    return x**2 - 3

def sinusoidal_function(x):
    """3. Sine Wave: y = sin(x). Used for modeling cycles, waves, and periodic data (e.g., sound)."""
    return np.sin(x)

def logarithmic_function(x):
    """4. Logarithm: y = log(x). Used when growth slows down over time (e.g., compound interest, very large scales)."""
    return np.log(x)

# --- Additional Core Mathematical Functions (9-13) ---

def absolute_value_function(x):
    """9. Absolute Value: y = |x|. Measures distance from zero (never negative)."""
    return np.abs(x)

def cubic_function(x):
    """10. Cubic Curve: y = x^3. Appears in slope, motion, and transformation contexts."""
    return x**3

def square_root_function(x):
    """11. Square Root: y = √x. Used where growth slows with magnitude (e.g., physics intensity)."""
    return np.sqrt(x)

def exponential_growth_function(x):
    """12. Exponential Growth: y = e^x. Models rapid increase (e.g., population, data scaling)."""
    return np.exp(x)

def signum_function(x):
    """13. Signum Function: y = -1 (x<0), 0 (x=0), 1 (x>0). Used to detect sign direction."""
    return np.sign(x)

# --- AI/ML Specific Functions (5-8) ---

def sigmoid_function(x):
    """5. The 'S' curve: y = 1 / (1 + e^-x). Used to convert any number into a probability (0 to 1)."""
    return 1 / (1 + np.exp(-x))

def relu_function(x):
    """6. Rectified Linear Unit: y = max(0, x). Modern and widely used Activation Function."""
    return np.maximum(0, x)

def tanh_function(x):
    """7. Hyperbolic Tangent: S-curve centered at zero. Squashes values between -1 and 1."""
    return np.tanh(x)

def gaussian_function(x, mu=0, sigma=2):
    """8. The Bell Curve: Used to describe natural distributions (like heights or errors)."""
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)


# --- 3. Visualization ---
# We now use a 7x2 grid to show all 13 functions clearly.

fig, axes = plt.subplots(7, 2, figsize=(14, 28)) # Updated size for 14 slots
axes = axes.flatten() # Flatten the 2D array of axes for easy indexing

# Title for the entire figure
fig.suptitle("The Building Blocks of AI/ML: 13 Essential Mathematical Functions", fontsize=18, y=0.995)

# Helper function to plot consistently (adapted from your input)
def plot_function(ax, func, x_domain, label, title, text_pos, text_desc, color, ylim=None, xlabel=''):
    ax.plot(x_domain, func(x_domain), label=label, color=color, linewidth=3)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.grid(True, linestyle=':', alpha=0.5)
    if ylim:
        ax.set_ylim(ylim)
    # Add explanatory text inside the plot
    ax.text(*text_pos, text_desc, fontsize=9, bbox=dict(facecolor='white', alpha=0.7))
    # Set Y-label only for the first column
    if ax.get_subplotspec().is_first_col():
        ax.set_ylabel('Output (y)')


# --- Plot 1: Linear Function (Index 0) ---
plot_function(axes[0], linear_function, xx, 'Linear: $y=2x+1$',
              '1. Linear Function (Direct Relationship)',
              (-9, 15),
              "**Purpose:** Shows a direct, steady relationship.\n**AI/ML Use:** **Linear Regression** (Simple Prediction).\nIt's the most basic way to guess a value based on input.",
              color='C0', ylim=(-20, 20))

# --- Plot 2: Quadratic Function (Index 1) ---
plot_function(axes[1], quadratic_function, xx, 'Quadratic: $y=x^2-3$',
              '2. Quadratic Function (Error/Loss Curve)',
              (-9, 60),
              "**Purpose:** Creates a 'U' shape to find the absolute lowest point.\n**AI/ML Use:** **Loss Functions** (Mean Squared Error).\nThe bottom of the 'U' represents the minimum error (best result).",
              color='C1', ylim=(-5, 80))

# --- Plot 3: Sinusoidal Function (Index 2) ---
plot_function(axes[2], sinusoidal_function, xx, 'Sinusoidal: $y=\sin(x)$',
              '3. Sinusoidal Function (Cycles)',
              (-9, 0.5),
              "**Purpose:** Models repeating up-and-down patterns.\n**AI/ML Use:** **Time Series Analysis**.\nUsed when data (like seasonality or waves) repeats over time.",
              color='C6', ylim=(-1.5, 1.5))
axes[2].axhline(0, color='gray', linestyle='--', alpha=0.6)

# --- Plot 4: Logarithmic Function (Index 3) ---
plot_function(axes[3], logarithmic_function, xx_pos, 'Logarithmic: $y=\\ln(x)$',
              '4. Logarithmic Function (Slowing Growth)',
              (1, 1.5),
              "**Purpose:** Changes fast initially, then slows dramatically.\n**AI/ML Use:** **Feature Scaling / Loss Calculation**.\nHelps manage very large differences in numbers (scales data) and is key in Cross-Entropy loss.",
              color='C7', ylim=(-3, 3))
axes[3].axhline(0, color='gray', linestyle='--', alpha=0.6)

# --- Plot 5: Sigmoid Function (Index 4) ---
plot_function(axes[4], sigmoid_function, xx, 'Sigmoid: $y=1/(1+e^{-x})$',
              '5. Sigmoid Function (Probability Converter)',
              (-9, 0.85),
              "**Purpose:** Squashes any input number into a value between 0 and 1.\n**AI/ML Use:** **Activation Function (Binary Classification)**.\nUsed to convert a neuron's output into a **probability** (e.g., Yes/No chance).",
              color='C2', ylim=(-0.1, 1.1))
axes[4].axhline(0.5, color='gray', linestyle='--', alpha=0.6)

# --- Plot 6: ReLU Function (Index 5) ---
plot_function(axes[5], relu_function, xx, 'ReLU: $y=max(0, x)$',
              '6. ReLU Function (Modern Activation Switch)',
              (-9, 8),
              "**Purpose:** A simple switch: If input is positive, pass it; if negative, output zero.\n**AI/ML Use:** **Activation Function (Deep Learning)**.\nIt's popular because it's fast and effective, helping deep networks train quickly.",
              color='C3', ylim=(-1, 15))

# --- Plot 7: Tanh (Hyperbolic Tangent) Function (Index 6) ---
plot_function(axes[6], tanh_function, xx, 'Tanh: $y=\\tanh(x)$',
              '7. Tanh Function (Centered S-Curve)',
              (-9, 0.7),
              "**Purpose:** Squashes values between -1 and 1.\n**AI/ML Use:** **Activation Function (Recurrent Networks)**.\nBeing centered at zero often helps models that process sequences (like text or speech) train more stably.",
              color='C5', ylim=(-1.1, 1.1))
axes[6].axhline(0, color='gray', linestyle='--', alpha=0.6)

# --- Plot 8: Gaussian Function (Index 7) ---
plot_function(axes[7], gaussian_function, xx, 'Gaussian (Bell Curve)',
              '8. Gaussian Function (Probability)',
              (-9, 0.15),
              "**Purpose:** Describes how data is naturally spread around an average.\n**AI/ML Use:** **Statistical Modeling / Probability Density**.\nUsed to calculate the likelihood of observing a particular data point.",
              color='C4', ylim=(-0.05, 0.25))

# --- Plot 9: Absolute Value Function (Index 8) ---
plot_function(axes[8], absolute_value_function, xx, 'Absolute Value: $y=|x|$',
              '9. Absolute Value Function (Distance)',
              (-9, 8),
              "**Purpose:** Measures how far a value is from zero (distance).\n**AI/ML Use:** **Loss Functions (L1 Loss)**, Feature Normalization.",
              color='C8', ylim=(-2, 10))
axes[8].set_xlabel('Input (x)') # Start of new row, add x-label

# --- Plot 10: Cubic Function (Index 9) ---
plot_function(axes[9], cubic_function, xx, 'Cubic Function: $y=x^3$',
              '10. Cubic Function (Symmetric Curvature)',
              (-9, 300),
              "**Purpose:** Shows both positive and negative curvature (S-shape).\n**AI/ML Use:** Polynomial regression, non-linear effects.",
              color='C9', ylim=(-1200, 1200))
axes[9].set_xlabel('Input (x)')

# --- Plot 11: Square Root Function (Index 10) ---
plot_function(axes[10], square_root_function, xx_pos, 'Square Root: $y=\\sqrt{x}$',
              '11. Square Root Function (Diminishing Returns)',
              (1, 2),
              "**Purpose:** Models diminishing progression as input grows.\n**AI/ML Use:** Activation or normalization in feature scaling.",
              color='C10', ylim=(-1, 4))
axes[10].set_xlabel('Input (x)')

# --- Plot 12: Exponential Growth Function (Index 11) ---
plot_function(axes[11], exponential_growth_function, xx, 'Exponential: $y=e^x$',
              '12. Exponential Function (Rapid Growth)',
              (-9, 5000),
              "**Purpose:** Represents rapid or compounding change.\n**AI/ML Use:** Weight updates, scaling in optimizers (Adam).",
              color='C11', ylim=(-500, 22000))
axes[11].set_xlabel('Input (x)')

# --- Plot 13: Signum Function (Index 12) ---
plot_function(axes[12], signum_function, xx, 'Signum Function: $y=sign(x)$',
              '13. Signum Function (Direction Detector)',
              (-9, 0.5),
              "**Purpose:** Identifies whether input is negative, zero, or positive.\n**AI/ML Use:** Gradient direction, control flows, input control.",
              color='C12', ylim=(-1.5, 1.5), xlabel='Input (x)')
axes[12].axhline(0, color='gray', linestyle='--', alpha=0.6)

# --- Final Layout Adjustments ---

# Hide the last unused subplot (Index 13)
axes[13].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.985]) # Adjust layout to make room for suptitle
plt.show()
```

# Derivatives

## forward Difference Derivatives
$f'(x) =(f(x+h)-f(x))/h$
- Not accurate as central Diffrence
- useful for simple tasks

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2**x
def true_derivative(x):
    return 2*x

def forward_diff(f,x,h=1e-5):
    return (f(x+h)-f(x))/h

x=5

y_forward=forward_diff(f,x)
print(y_forward)

plt.figure(figsize=(10,7))
plt.plot(x,y_forward,label='Forward Diffrence Derivative')
plt.title('Numerical Derivative Representation')
plt.grid(True)
plt.legend()
plt.plot()
```

## Backward Diffrence Derivative
$f'(x)=(f(x)-f(x-h))/h$<br>
**Use**
- You only know values behind the point (x - h)
- End of a dataset (last data point)
- Data comes in reverse order

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2**x
def prime_f(x):
    return 2*x
def backward_diff(f,x,h=1e-5):
    return (f(x)-f(x-h))/h
    

x=10
y_backward=backward_diff(f,x)
print(y_backward)

plt.figure(figsize=(10,8))
plt.plot(x,y_backward,label='Backward Diffrenece Derivative')
plt.title('Numerical Backward Diff Derivative')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Central Diffrence Derivative
$f'(x)=(f(x+h)-f(x-h))/2h$<br>
**Use**
- You want high accuracy
- Mathematics, physics, machine learning
- Function is known at all points

```python
import numpy as np
import matplotlib.pyplot as plt

def f3(x):
    return x**2
def p_f2(x):
    return 2*x
def central_diff(f3,x,h=1e-5):
    return (f3(x+h)-f3(x-h))/(2*h)

x=np.linspace(-5,5,400)
y_central=central_diff(f3,x)
plt.figure(figsize=(10,8))
plt.plot(x,y_central,label='Central Diffrence Derivative',color='green')
plt.title('Numerical Representration of Central Derivative',color='red')
plt.grid(True,alpha=0.5,linestyle='--')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f4(x):
    return 2**x
def p_f3(x):
    return 2*x
def second_der(f4,x,h=1e-5):
    return (f4(x+h)-2*f4(x)+f4(x-h))/(h**2)

x=np.linspace(-5,5,400)
y_second=second_der(f4,x)
y_true=p_f3(x)
plt.figure(figsize=(10,8))
plt.plot(x,y_true,'--',label='True Derivative(2x)',color='red')
plt.plot(x,y_second,label='Second Derivative',color='green')
plt.title('Numerical Representation',color='black')
plt.grid(True,alpha=0.5,linestyle='--',color='violet')
plt.legend()
plt.tight_layout()
plt.show()
```

## Differentiation

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-2,2,100)
y=x**3-3*x**2+2*x
dy_dx=3*x**2-6*x+2

plt.plot(x,y,'--',label='$x^3-3x^2+2x$',color='violet')
plt.plot(x,dy_dx,label='$3x^2-6x+2$',color='grey')

# For Tangent
x_v=1
y_v=x_v**3-3*x_v**2+2*x_v
slope=3*x_v**2-6*x_v+2
tangent_x=np.linspace(x_v-0.5,x_v+0.5,10)
tangent_y=y_v+slope*(tangent_x-x_v)
plt.plot(tangent_x,tangent_y,'r-',label='Tangent line at X=1')
plt.scatter([x_v],[y_v],color='red')
plt.text(x_v+0.1,y_v,'Point (1, f(1))',color='red')

plt.xlabel('F(x)')
plt.ylabel('dy/dx')
plt.title('Function with Derivative',fontweight='bold',fontsize=15)
plt.grid(True,alpha=0.5,linestyle='--')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
xDomain=[-2,2]
numSteps=41
x=np.linspace(xDomain[0],xDomain[1],numSteps)
y1=np.exp(x)
y2=np.exp(x**2)
y3=np.exp((-x)**2)
y4=np.exp(-(x**2))
y5=np.exp(x)**2

plt.figure(figsize=(8,8))
plt.plot(x,y1,linewidth=2,label='exp(x)')
plt.plot(x,y2,linewidth=2,label='$exp(x^2)$')
plt.plot(x,y3,'-',linewidth=2,label='$exp((-x)^2)$')
plt.plot(x,y4,'--',linewidth=2,label='$exp(-(x^2))$')
plt.plot(x,y5,linewidth=2,label='$exp(x)^2$')
plt.xlim([x[0],x[-1]])
plt.ylim([-1,10])
plt.xlabel('X')
plt.ylabel('Exponential Function')
plt.legend()
plt.grid(True)
plt.show()
```
