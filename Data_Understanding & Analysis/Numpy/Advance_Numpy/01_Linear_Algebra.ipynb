### 1.1 Euclidean Vector

- **where we used to measure the distance in 2D or 3D**

```python
# example 1.
import numpy as np
import matplotlib.pyplot as plt

pos1=np.array([4,7])
pos2=np.array([5,3])

plt.figure(figsize=(8,8))

plt.quiver(0,0,pos1[0],pos1[1],scale_units='xy',angles='xy',scale=1,color="black",label="Position(4,7)")
plt.quiver(0,0,pos2[0],pos2[1],scale_units='xy',angles='xy',scale=1,color="red",label="Position(5,3)")
plt.xlim(0,8)
plt.ylim(0,8)
plt.title("Geometric Representation of Co-ordinates",color="green",fontweight="bold",fontsize=15)
plt.grid(True)
plt.legend()
plt.show()
```

### 1.2 Vector Addition and Subtrcation

```python
import numpy as np 
import matplotlib.pyplot as plt

v1=np.array([5,3])
v2=np.array([3,4])

# addition
add=v1+v2

# subtraction
sub=v2-v1

plt.figure(figsize=(7,7))
plt.quiver(0,0,v1[0],v1[1],scale_units="xy",angles="xy",scale=1,color="orange",label="V1(2,3")
plt.quiver(v1[0],v1[1],v2[0],v2[1],scale_units="xy",angles="xy",scale=1,color="yellow",label="V2(3,4)")
plt.quiver(0,0,add[0],add[1],scale_units="xy",angles="xy",scale=1,color="red",label="Addition")
plt.quiver(0,0,sub[0],sub[1],scale_units="xy",angles="xy",scale=1,color="blue",label="Substraction")
plt.xlim(-2,8)
plt.ylim(-2,8)
plt.title("Geometric Representation of Co-cordinates",color="blue",fontweight="bold",fontsize=15)
plt.grid(True)
plt.legend()
plt.show()
```

### 1.3 Vector Independence/Dependence


```python
# example 1.
import numpy as np

a=np.array([1.5,3])
b=np.array([3,6])
matrix=np.column_stack((a,b))
rank=np.linalg.matrix_rank(matrix)
print(matrix)
print(rank)
```

```python
import numpy as np

v1 = np.array([1, 2])
v2 = np.array([2, 4])  # Dependent, 2*v1
matrix = np.column_stack((v1, v2))
rank = np.linalg.matrix_rank(matrix)
print(matrix)
print("Rank:", rank)  # 1, dependent
```

### Norm and Normalisation of Vector

```python
# example 1.
import numpy as np
v_1=np.array([5,6])
norm1=np.linalg.norm(v_1)
normalisation1=v_1/norm1

print("Norm:",norm1)
print("Normalisation:",normalisation1)
```

```python
# example 2.
import numpy as np
import matplotlib.pyplot as plt
v_2=np.array([2,3])
norm2=np.linalg.norm(v_2)
normalisation2=v_2/norm2

print("Norm:",norm2)
print("Normalisation:",normalisation2)

plt.figure(figsize=(7,7))
plt.quiver(0,0,v_2[0],v_2[1],angles="xy",scale_units="xy",scale=1,color="blue",label="V_2(2,3)")
plt.quiver(0,0,normalisation2[0],normalisation2[1],angles="xy",scale_units="xy",scale=1,color="orange",label="Normalisation(0.5547002 ,0.83205029)")
plt.xlim(0,3)
plt.ylim(0,4)
plt.title("Geometric Representation")
plt.grid(True)
plt.legend()
plt.show()
```
