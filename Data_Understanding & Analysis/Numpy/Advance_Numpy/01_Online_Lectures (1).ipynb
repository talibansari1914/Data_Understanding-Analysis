 - set operations

```python
set_1={1,2,3,4,5,6,7}
set_2={5,6,4}
# Union
un=set_1 | set_2
un2=set_1.union(set_2)
print("Union1:",un)
print("Union2:",un2)
```

```python
# Intersection (Common)
in1=set_1 & set_2
in2=set_1.intersection(set_2)
print("Intersection1:",in1)
print("Intersection2:",in2)
```

```python
# Symmetric difference
diff=set_1.difference(set_2)
diff2=set_1.symmetric_difference(set_2)
print("Diff:",diff)
print("Diff2:",diff2)
```

```python
# Disjoint
dis1=set_1^set_2

print("Dis1:",dis1)
```

```python
# Subset
print(set_2.issubset(set_1))
```

```python
# Superset
print(set_1.issuperset(set_2))
```

```python
# Exclusive or Ordinary progrmas of Set
# 1.
# set_1={1,2,3,4,['a','b']}
# print("Set_1:",set_1)
# print("---"*20)

set_2={'a','b','c',(2,3)}
print("Set_2:",set_2)
print("---"*20)

# set_3={1,'a','c',{1,2}}
# print("Set_3:",set_3)
```

```python
try:
    set_1={1,2,3,4,['a','b']}
    print("Set_1:",set_1)
except TypeError as t:
    print(f"Set Can not be written in nested form")
```

```python
try:
    set_3={1,'a','c',{1,2}}
    print("Set_3:",set_3)
except TypeError as t:
    print(f"Error:{t}")
```

```python
set_4={1,1,1,1,1,0,0,0,0}
print("Set 4:",set_4)
print("---"*30)

set_5={1,1,1,1,1,1,True}
print("Set 5:",set_5)
print("---"*30)


set_6={True,1,1,1,1,1}
print("Set 6:",set_6)
print("---"*30)


set_7={1,True,True,True,True,True,True}
print("Set 7:",set_7)
print("---"*30)


set_8={0,False ,False,0,0,False ,0,0}
print("Set 8:",set_8)
```
