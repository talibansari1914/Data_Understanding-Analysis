```python
# PROBLEM.1
Task 1.1: Create a 1-D NumPy array named daily_revenue_lakhs from the provided list. The values are in Lakhs INR.
Task 1.2: Calculate and print the total revenue for the two weeks.
Task 1.3: Calculate and print the average daily revenu
```

```python
import numpy as np
daily_revenue_data = [
5.5, 6.2, 5.8, 7.1, 7.5, 8.0, 9.2,
6.0, 6.5, 7.0, 7.8, 8.5, 9.0, 9.5]
# List to Numpy Array Conversion
daily_revenue_lakhs=np.array(daily_revenue_data)
```

```python
# Total Revenue Calculation
total_revenue=daily_revenue_lakhs.sum()
print(f"Total Revenue:{total_revenue}")
```

```python
# Average daily revenue Calculation
avg_daily_revenue=daily_revenue_lakhs.mean()
print("Average Daily Revenue :",avg_daily_revenue)
```

```python
# PROBLEM.3
Task 3.1: Create a 1-D NumPy array named unit_costs with the provided data.
Task 3.2: Define a variable profit_margin_percent with a value of 25.
Task 3.3: Calculate the selling price for each unit by adding the profit margin. The formula is: selling_price =
unit_cost + (unit_cost * profit_margin_percent / 100). Store the result in a new array named selling_prices.
```

```python
# PROBLEM.4
Task 4.1: Create a 2-D NumPy array named ingredients_used. The rows represent three flavors and the columns
represent three ingredients (e.g., 'Vanilla', 'Chocolate', 'Fruits').
Task 4.2: Create a 1-D NumPy array named batches_produced representing the number of batches for each flavor.
Task 4.3: Use NumPy broadcasting to calculate the total amount of each ingredient used across all batches. Print the result
```

```python
import numpy as np
# Given data
# rows: Flavors (A, B, C)
# cols: Ingredients (Vanilla, Chocolate, Fruit)
ingredients_used_per_batch = [
[0.5, 0.3, 0.1],
[0.2, 0.6, 0.0],
[0.4, 0.2, 0.3]
]
batches_produced_data = [100, 150, 200]
# 2D Array Creation
ingredients_used=np.array(ingredients_used_per_batch)
print(f"Ingredients Used :{ingredients_used}")
batches_produced=np.array(batches_produced_data)
print(f"Batches Produced:{batches_produced}")
```

```python
amount_1=ingredients_used*batches_produced
total_amount_ingredients=amount_1.sum(axis=0)
print("Total Amount:",total_amount_ingredients)
```

```python
# Each Product's Total Amount
vanilla,chocolate,fruit=total_amount_ingredients
print("Total Amount:")
print("Vanilla Price:",vanilla)
print("Chocolate Price:",chocolate)
print("Fruit Price:",fruit)
```
