## Advanced function
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

np.dot(A, B)       # matrix multiplication
A @ B              # same as dot
np.transpose(A)    # transpose
np.linalg.inv(A)   # inverse
np.linalg.det(A)   # determinant
np.linalg.eig(A)   # eigenvalues, eigenvectors
np.linalg.solve(A, np.array([1,2]))  # solve Ax = b
## random function
np.random.rand(3,2)       # uniform [0,1)
np.random.randn(3,2)      # normal distribution
np.random.randint(0,10,5) # 5 integers between 0-9
np.random.choice([1,2,3]) # random element
np.random.seed(42)        # reproducibility
## flatten
a = np.arange(12)       # [0..11]
a.reshape(3,4)          # 3x4 matrix
a.ravel()               # flatten
a.flatten()             # flatten copy
a.T                     # transpose
## sorting and indices
a = np.array([3,1,2])
np.sort(a)         # [1,2,3]
a.argsort()        # [1,2,0] (indices)

b = np.array([[1,4],[3,1]])
np.sort(b, axis=0) # sort column-wise

np.where(a > 1)    # indices where condition true
np.argmax(a)       # index of max
np.argmin(a)       # index of min

## identity
np.identity(3)       # identity matrix
np.diag([1,2,3])     # diagonal matrix
np.tri(3)            # lower triangular
