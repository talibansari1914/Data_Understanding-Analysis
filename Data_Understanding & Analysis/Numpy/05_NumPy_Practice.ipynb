```python
import numpy as np
a1=np.array([[10,20,3],[11,12,4]])
a2=np.array([[5,10,1],[15,20,2],[25,30,3]])
print("dot product:\n",np.dot(a1,a2))
t=np.transpose(a1)
print("Before transpose:\n",a1)
print("After transpose:\n",t)
```

```python
inv=np.linalg.inv(a2)
print("Before inv:\n",a2)
print("After inv:\n",inv)
```

```python
det=np.linalg.det(a2)
print("Before determinant:\n",a2)
print("After determinant:\n",det)
```

```python
eigenvalue ,eigenvactor=np.linalg.eig(a2)
print("Eigen value:\n",eigenvalue)
print("Eigen vactor:\n",eigenvactor)
```
