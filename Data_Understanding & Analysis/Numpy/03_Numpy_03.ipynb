```python
import numpy as np
a=np.array([[5,10],[20,25]])
b=np.array([[2,3,4],[5,7,9]])
print("\ndot product:\n",np.dot(a,b))
inv=np.linalg.inv(a)
print("\nInverse of a matrix:\n",inv)
det=np.linalg.det(a)
print("\nDeterminant:\n",det)
eigenvalue , eigenvector=np.linalg.eig(a)
print("\nEigenvalue:\n",eigenvalue)
print("\nEigenvector:\n",eigenvector)
```
